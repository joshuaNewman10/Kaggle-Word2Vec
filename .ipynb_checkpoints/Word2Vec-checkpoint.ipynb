##Get and Clean Data

```python
#read data from file
unlabeledTrainingData = pd.DataFrame(readFile('data/unlabeledTrainData.tsv', 0, '\t', 3))
labeledTrainingData = pd.DataFrame(readFile('data/labeledTrainData.tsv', 0, '\t', 3))
testingData = pd.DataFrame(readFile('data/testData.tsv', 0, '\t', 3))
```

Cleaning will be similar to what we did in the Bag of Words verison. But here we won't be removing stop words since they will be useful to our Words2Vec Deep Learning approach. This algorithm actually relies on the broader context of the sentence to produce high quality word vectors. Thus make stop word removal optional in our functions below. Good to make another verison where we dont remove numbers!

We want specific input format here (sentences which are word lists). Difficult to do because english sentences can end with lots of stuff and punctuation/capitalization isnt a good guide either. We will use the NLTK's punkt tokenizer for sentence splitting. 

```python
# Download the punkt tokenizer for sentence splitting
import nltk.data
nltk.download()   

# Load the punkt tokenizer
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
#load up our reviewToSentences func
from reviewParser import reviewToSentences
```

```python
#Parse the raw reviews into a list of sentences which themselves are list of words
sentences = []
print('parsing sentences from labeled training set: \n')
for review in labeledTrainingData['review']:
    sentences += reviewToSentences(review, tokenizer, False)
    
print('parsing sentences from unlabled training set: \n')
for review in unlabeledTrainingData['review']:
    sentences += reviewToSentences(review, tokenizer)
```

```python
sentences
```

```python
# Import the built-in logging module and configure it so that Word2Vec 
# creates nice output messages
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',\
    level=logging.INFO)

# Set values for various parameters
num_features = 300    # Word vector dimensionality                      
min_word_count = 40   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size                                                                                    
downsampling = 1e-3   # Downsample setting for frequent words

# Initialize and train the model (this will take some time)
from gensim.models import word2vec
print "Training model..."
model = word2vec.Word2Vec(sentences, workers=num_workers, \
            size=num_features, min_count = min_word_count, \
            window = context, sample = downsampling)

# If you don't plan to train the model any further, calling 
# init_sims will make the model much more memory-efficient.
model.init_sims(replace=True)

# It can be helpful to create a meaningful model name and 
# save the model for later use. You can load it later using Word2Vec.load()
model_name = "300features_40minwords_10context"
model.save(model_name)
```

###From Words to Paragraphs Attempt 1:  Vector Averaging

Challenge for ou DS is that we need to deal with variable-length reviews. Gotta find way to take individual word vectors and transform them into feature set that is same length for every review.

Each word is in 300 dimensional vector space so we can use vector operations to combine the words in each review.

Method 1: Just average the word vectors in a given review (good to remove stop words here since will just add noise)


```python
from reviewParser import reviewToWordList
from reviewParser import getAvgFeatureVectors
```

```python
from reviewParser import getAvgFeatureVectors
```

```python
#Calculate average feature vector for training and test sets using out reviewParser utility functions
#Note we are now using stop word removal

print('creating average feature vectors for labeled training reviews')

cleanedTrainingReviews = []
for review in labeledTrainingData['review']:
    cleanedTrainingReviews.append(reviewToWordList(review, removeStopWords=True))
                            
trainingDataVectors = getAvgFeatureVectors(cleanedTrainingReviews, model, num_features)

print('creaitng average feature vectors for testing reviews')

cleanedTestingReviews = []
for review in testingData['review']:
    cleanedTestingReviews.append(reviewToWordList(review, removeStopWords=True))
    
testingDataVectors = getAvgFeatureVectors(cleanedTestingReviews, model, num_features)
                                  
```

####Train Random Forest on Training Average Feature Vectors

```python
#Fix RF to training data using 100 trees
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100)

print('fitting a random forest to our labled training data')
forest = forest.fit(trainingDataVectors, labeledTrainingData['sentiment'])
```

```python
#Test forest on testing data and extract out our results
vectorAvgResults = forest.predict(testingDataVectors)

#Write the test results
output = pd.DataFrame(data={'id':testingData['id'], 'sentiment': vectorAvgResults})
outputCSVFile(output, 'Word2Vector_avgVectors', index=False, quoting=3)
```

This model produces better result than random chance but actually underperforms our original model (Bag of Words) by a few percentage points.

So element-wise average didn't do great. Better way? Another approach is to apply tf-idf weights. This measures how 'important' a word is within a given set of documents.
