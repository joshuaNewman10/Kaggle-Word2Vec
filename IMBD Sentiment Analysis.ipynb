Porter Stemming and Lemmatizing (both available in NLTK) would allow us to treat "messages", "message", and "messaging" as the same word, which could certainly be usefu

```python
#clean out the whole training dataset
numReviews = len(trainingReviews)
cleanedReviews = []
```

```python
from reviewParser import reviewToWords
#loop over each review, clean and append
for index in xrange(0, numReviews):
    #print message when done with 1000 reviews
    if((index+1)%1000 == 0):
        print("Review %d of %d\n" % ( index+1, numReviews ))
    cleanedReviews.append( reviewToWords(trainingReviews.ix[index]['review']))
```

###Create Features From Bag of Words

```python
print('creating Bag of Words features...')
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Initialize  CountVectorizer Object (sckits BoW tool) which will make sparse representation
vectorizer = CountVectorizer(analyzer = 'word',   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features =5000)
```

Note that count vectorizer comes with its own tokenizing, preprocessing and stop word functionality. We could have used these internal methods instead of doing it ourself, but wheres the fun in that!

fit_transform() will:
- fit modeland learn the vocab
- transforms training data into feature vectors

I/O
- Input: list of strings
- Output: sparse feature representation

```python
trainingDataFeatures = vectorizer.fit_transform(cleanedReviews)
```

```python
#Use numpy array bc easier to work with
trainingDataFeatures = trainingDataFeatures.toarray()
```

```python
#25k obs by 5k features (one for each vocab word --> set UB of 5k)
print('shape of training data features: ',trainingDataFeatures.shape )
```

```python
#Explore words in our new vocabulary
reviewVocab = vectorizer.get_feature_names()
print('reviewVocab', reviewVocab)
```

```python
#Print count of vocab words
featureDistribution = np.sum(trainingDataFeatures, axis=0)

#iterate over featureDistribution and print vocab word with its count
for feature, count in zip(reviewVocab, featureDistribution):
    print(feature, count)
```

##Random Forest Analysis

We now have numeric training features from BoW and the original sentiment labels for each of our feature vectors. Time for supervised learning.

```python
print('Training the random forest')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#Initialize RF with default of 100 trees
reviewForest = RandomForestClassifier(n_estimators = 100)
```

```python
#Fit reviewForest to training set
reviewForest = reviewForest.fit(trainingDataFeatures, trainingReviews['sentiment'])
```

##Create Submission

```python
testingReview = pd.DataFrame(readFile('data/testData.tsv', 0, '\t', 3))
```

```python
#Verify our shape matches the trained forest
print('shape of testingReviews data: ', testingReview.shape)
```

###Preprocess Testing Data

```python
numberTestingReviews = len(testingReview['review'])
cleanedTestReviews = []
```

```python
print('cleaning and parsing testing reviews...\n')
#note xrange uses generators internally
for index in xrange(0, numberTestingReviews):
    if( (index+1)%1000 == 0 ):
        print('Review of %d of %d\n' % (index+1, numberTestingReviews))
    cleanedTestReviews.append(reviewToWords(testingReview.ix[index]['review']))
```

```python
#Get Bag of Words for test set, convert to numpy arrray
testDataFeatures = vectorizer.transform(cleanedTestReviews)
testDataFeatures = testDataFeatures.toarray()
```

###Use RF To Make Predictions

```python
reviewResults = reviewForest.predict(testDataFeatures)
```

```python
#copy results to pandas df with an id collumn and sentiment columnn (expected submission format)
output = pd.DataFrame( data = {'id': testingReview['id'] , 'sentiment': reviewResults})
```
